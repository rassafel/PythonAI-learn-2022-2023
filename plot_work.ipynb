```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

ax = fig.add_subplot()

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot([0,1,2,3,4,5], [5,4,3,2,1,0])

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot([0,1,2,3,4,5], [5,4,3,2,1,0], marker='o', linestyle='dashed')
ax.scatter([1,2,3], [3,4,1], marker="*", color='red')

plt.show()
```

```python
df = {'jan': {'x': [1,2,3,4], 'y': [4,3,2,1]}, 
      'dec': {'x': [8,7,6,5], 'y': [5,6,7,8]}}

for key, item in df.items():
    plt.plot(item['x'], item['y'], label=key)

plt.legend()
plt.show()
```

```python
df = {'jan': 10, 'dec': 20}
    
plt.bar('jan', df['jan'])
plt.bar('dec', df['dec'])
```
